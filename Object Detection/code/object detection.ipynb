> ## `Importing Pre-Req Packages`

```python
# using Image AI pckg for Object detection
from imageai.Detection import ObjectDetection

# importing OS  for accessing files in the system
import os

# importing timeit  for calculating ELAPSED TIME i.e.( time of Execution of a model)
import timeit

# Matplotlib is a plotting library for the Python programming language
# for Visualization
import matplotlib.pyplot as plt


# numpy adding support for large, multi-dimensional arrays and matrices
import numpy as np
import cv2 as img
```

> ## `Retina Net Algo.`

```python
detector = ObjectDetection()

detector.setModelTypeAsRetinaNet()

execution_path = r"C:\Users\venka\PROJECT"

detector.setModelPath( os.path.join(execution_path , "resnet50_coco_best_v2.0.1.h5"))

detector.loadModel()

# timeit.default_timer() is used instead of time.time() becoz it works on any platform

start_time = timeit.default_timer()

detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path , "test-input4.jpg"), output_image_path=os.path.join(execution_path , "T_Output11.jpeg"))
obj1 = 0
acc1 = 0
for eachObject in detections:
    print(eachObject["name"] , " : " , eachObject["percentage_probability"] )
    obj1 = obj1 + 1
    acc1 += eachObject["percentage_probability"]

# elapsed time is the time taken from "Detecting Objects in an image" TO "Printing each Object with Accuracy"
elapsed = timeit.default_timer() - start_time

# just printing the Average of all Objects Accuracy ( simply ---> [ sum of Accuracies// no. of objects] )
accu1 = float(acc1/obj1)
print("elapsed time for RETINA NET is : ",elapsed)
print("Objects detected : ",obj1," avg acc : ",accu1)
```

```python

# "imread() --> To Read an image from a file into an array.
im = img.imread(r'C:\Users\venka\PROJECT\T_output11.jpeg') 

# plt --> matplotlib.Pyplot
# rc -->  used to customize all kinds of properties in matplotlib (Run,Configure)
plt.rcParams["figure.figsize"] = (13,13)

# imshow()  --> Display the Image
plt.imshow(im)
```

> ## `Comparison of Alogorithm using BarGraph`

```python

labels = ['Yolo_v3','Retina Net', 'Yolo_Tiny']
obj_count = [obj1,obj2,obj3]
Avg_accuracy = [accu1,accu2, accu3]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, obj_count, width, label='No. of Objects detected')
rects2 = ax.bar(x + width/2, Avg_accuracy, width, label='avg. Accuracy of all Objects')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('count of objs & Accuracy')
ax.set_title('Comparison of Algorithms')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```
